#### QN 5. For any column involving strings, find all the unique values and print them out. Once you have identified the distinct values, assign a number starting from 0 to the max count of unique values.

```python
data["type"].unique()
```

```python
import pandas as pd
data = pd.read_csv("netflix_titles.csv")
var=data['type'].unique()
dic={}
count=0
size=len(var)
for a in var:
    if count<size:
        dic[a]=count
        count=count+1
        
dic
```

#### QN 7. For all the numerical columns, normalise the data by using the normalise formula (squash the values between 0 and 1). Use information in the lecture. 

```python
data.head()
```

```python
def normalise(col):
    return (col - col.min()) /(col.max() - col.min())#formula

release_year = np.array(data["release_year"].values)

#normalise
release_year_normalise = normalise(release_year)
release_year_normalise
```

#### QN 9. On canvas there will be a CSV file called “quaddata.csv.” Your program will solve quadratic equations. The file will give you the a, b and c values. You will need to research on how to use the quadratic formula if you have not done so before. NOTE – some of the data will give you 2 roots, 1 root and no roots. Your program should offer the service of giving all three results depending on the input values.


```python
import cmath
def calcEquations(a,b,c):
    #discriminant
    dis = (b**2)-(4*a*c)
    if dis>0:
        ans1 = (-b-cmath.sqrt(dis))/(2 * a)
        ans2 = (-b + cmath.sqrt(dis))/(2 * a)
        print(f"{ans1}{ans2}")
    elif dis==0:
        ans3=(-b) / 2*a
        print(f"{ans3}")
    else:
        print("No root for dis less than 0")
    
#import file
quad = pd.read_csv("quaddata.csv")
data = np.array(quad)
df = pd.DataFrame(data=data,columns=['a','b','c'])

for i in range(len(df)):
    a = df.a[i]
    b = df.b[i]
    c = df.c[i]
    calcEquations(a,b,c)
```
